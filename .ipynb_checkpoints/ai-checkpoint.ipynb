```python
import pandas as pd
import numpy as np #remove maybe
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from datetime import datetime
```

```python
sns.set_theme(style="darkgrid")
# read dataset
path = './melb_data (2).csv'
melb_data = pd.read_csv(path)


```

```python
melb_data
```

```python
melb_data.info()
```

```python
melb_data.dropna(inplace=True)
current_year = datetime.now().year

# Create a new column 'age' by subtracting 'yearbuilt' from the current year
melb_data['Age'] = current_year - melb_data['YearBuilt']
```

```python
melb_data.info()
```

```python

X=melb_data.drop(['Price'],axis=1).drop(['YearBuilt'],axis=1).drop(['Postcode'],axis=1)

y=melb_data['Price']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
train_data=X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data=train_data.select_dtypes(exclude=[object])
```

```python
train_data.corr()
```

```python
train_data['Bathroom']=np.log(train_data['Bathroom']+1)
train_data['Bedroom2']=np.log(train_data['Bedroom2']+1)
train_data['Landsize']=np.log(train_data['Landsize']+1)
train_data['Age']=np.log(train_data['Age']+1)
train_data['BuildingArea']=np.log(train_data['BuildingArea']+1)
train_data['Price']=np.log(train_data['Price']+1)
```
